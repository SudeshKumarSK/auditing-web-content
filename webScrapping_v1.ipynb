```python
# Indicates the number of times we call the getTaggedPostsUserData() method to get the user_id and blog name from the posts. Every time we run this while loop,
# 20 users' data will be collected.
numberTimes = 5
i = 0
timeStamp = None
blogNames = []
while(i < numberTimes):
    response = ""
   
    if i == 0:
        response = tumblrScraper.getTaggedPostsUserData(tagName=tagName, timestamp=None)
        if response["status"]:
            timeStamp = response["timestamp"]
            for key, blog in response["response"].items():
                blogNames.append(blog["blog_name"])
                
    else:
        print(timeStamp)
        response = tumblrScraper.getTaggedPostsUserData(tagName=tagName, timestamp=timeStamp)
        if response["status"]:
            timeStamp = response["timestamp"]
            for key, blog in response["response"].items():
                blogNames.append(blog["blog_name"])
                
    i += 1
```
